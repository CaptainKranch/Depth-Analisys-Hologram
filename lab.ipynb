```python
import pandas as pd
```

```python
df = pd.read_csv('data/holo.csv')
```

```python
value = ['Single Carrier', 'Undefined']

df = df[df.plan_type.isin(value) == False ]
```

```python
var = df['name'].value_counts()

df = df[df['name'].isin(var[var>1].index)]   

df['name'].value_counts()
```

```python
serial_numbers = df['name'].value_counts()

sn_size = serial_numbers.size

count=0

one_carrier_list = []

for _ in range(sn_size):
    controller_serial_number =  serial_numbers.keys()[_]
    
    Sdf = df.loc[df['name'].isin([controller_serial_number])]
    value = Sdf['network_name'].value_counts()
    value_size = value.size

    if value_size == 1:
        one_carrier_list.append(controller_serial_number)
        df = df[~((df['name'] == controller_serial_number))]
    else:
        count+=1
```

```python
df['timestamp'] = df['timestamp'].apply(pd.to_datetime)

df['timestamp'] = df['timestamp'].dt.date

df = df.drop(['session_begin', 'imei', 'sim'], axis = 1)

groupby_name_timestamp = df.groupby(['name','timestamp'], as_index=False)['network_name'].agg(['value_counts'])
```

```python
groupby_name_timestamp.to_csv('data/holo_groupby_name_timestamp.csv')
```

```python
df = pd.read_csv('data/holo_groupby_name_timestamp.csv')

date = df['timestamp'].value_counts()
date_size = date.size

delete = []
no_delete = []
date_list = []
count = 0

for _ in range(date_size):
    
    current_date = date.keys()[_]
    date_df = df.loc[df['timestamp'].isin([current_date])]

    sn = date_df['name'].value_counts()
    sn_size = sn.size

    for __ in range(sn_size):
        sn_count = sn[__]
        sn_var = sn.keys()[__]

        if sn_count == 1:
            delete.append(sn_var)
            date_list.append(current_date)
            df = df.loc[~((df['name'] == sn_var) & (df['timestamp'] == current_date)),:]
        else:
            no_delete.append(sn_var)
            count+=1
```
