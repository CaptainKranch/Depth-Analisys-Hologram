```python
import pandas as pd
```

```python
df = pd.read_csv('data/holo.csv')
```

```python
value = ['Single Carrier', 'Undefined']

df = df[df.plan_type.isin(value) == False ]
```

```python
var = df['name'].value_counts()

df = df[df['name'].isin(var[var>1].index)]   
```

```python
serial_numbers = df['name'].value_counts()

sn_size = serial_numbers.size

count=0

one_carrier_list = []

for _ in range(sn_size):
    controller_serial_number =  serial_numbers.keys()[_]
    
    Sdf = df.loc[df['name'].isin([controller_serial_number])]
    value = Sdf['network_name'].value_counts()
    value_size = value.size

    if value_size == 1:
        one_carrier_list.append(controller_serial_number)
        df = df[~((df['name'] == controller_serial_number))]
    else:
        count+=1
```

```python
df = df.drop(['timestamp', 'imei', 'sim'], axis = 1)

df['session_begin'] = df['session_begin'].apply(pd.to_datetime)

df['session_begin_date'] = df['session_begin'].dt.date

groupby_name_timestamp = df.groupby(['name','session_begin_date'], as_index=False)['network_name'].agg(['value_counts'])
```

```python
all_sns = timeline_df['name'].value_counts()
sns_size = all_sns.size

help_net = []
help_cont = []
sum_duration_list = []

first_net = timeline_df.iloc[0]
first_net = first_net[-1]
help_net.append(first_net)

first_cont = timeline_df.iloc[0]
first_cont = first_cont[1]
help_cont.append(first_cont)


duration_list = timeline_df['duration'].values
sn_list = timeline_df['name'].values

for _ in range(sns_size):
    curr_net = timeline_df.iloc[_ + 1]
    curr_net = curr_net[-1]

    curr_cont = timeline_df.iloc[_ + 1]
    curr_cont = curr_cont[1]

    if curr_net == help_net[-1] and curr_cont == help_cont[-1]:
        sum_duration = duration_list[_ + 1] + duration_list[_]
        sum_duration_list.append(sum_duration)
        help_net.append(curr_net)
        help_cont.append(curr_cont)
        owo = timeline_df.drop([timeline_df.index[_]])
    else:
        help_net.append(curr_net)
        help_cont.append(curr_cont)
        
```

```python
owo.drop(['duration'], axis=1)
owo['sum_duration'] = sum_duration_list
```
